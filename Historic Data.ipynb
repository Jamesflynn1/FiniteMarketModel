```python
import numpy as np
import numpy as np
import csv
import requests
from contextlib import closing
import csv
from codecs import iterdecode
```

```python
#Obtain current FTSE 100 tickers from a local file

file = open("FTSE100.tickers")
iter = csv.reader(file, delimiter='\n')
tickers = []
for ticker in iter:
    tickers=tickers+ticker
```

```python
#Obtain historical data from Yahoo finance
yearsOfData = 5
endTime = 1701561600

startTime = endTime- 24*60*60*365 *yearsOfData
tickerPlaceholder = "PLACEHOLDER"
webURL = f"https://query1.finance.yahoo.com/v7/finance/download/{tickerPlaceholder}?period1={str(startTime)}&period2={str(endTime)}&interval=1d&events=history&includeAdjustedClose=true"

priceData = {}
for stockTicker in tickers:
    requestURL = webURL.replace(tickerPlaceholder, stockTicker)
    firstRow = True
    with requests.get(requestURL, stream=True, headers={'User-agent': 'Mozilla/5.0'}) as r:
        lines = (line.decode('utf-8') for line in r.iter_lines())
        priceData[stockTicker] = []
        for row in csv.reader(lines):
            if firstRow:
                firstRow = False
            else:            
                priceData[stockTicker].append(row)
            if (row == ["404 Not Found: Timestamp data missing."] or row == ['404 Not Found: No data found', ' symbol may be delisted']):
                print (stockTicker)    
```

```python
import pandas as pd

for prices in list(priceData.keys()):
    print(prices)
    stock = pd.DataFrame((np.array(priceData[prices]))[:,5])
    stock.to_csv(f"HistoricPrices/{prices.replace(".L","")}.csv")
```
